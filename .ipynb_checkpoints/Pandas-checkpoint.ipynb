```python
#1. Create DF using dict:
import pandas as pd
from IPython.display import display
pd.set_option('display.max_colwidth', 40)
pd.set_option('display.precision', 5)
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 30)


dict_columns = {
    'contents':['Author', 'Book', 'Target', 'No_Donation'],
    'infos':['Pham Dinh Khanh', 'ML algorithms to Practice', 'Vi mot cong dong AI vung manh hon', 'Community'],
    'numbers':[1993, 2021, 1, 2]
}

df = pd.DataFrame(dict_columns)
display(df)
```

```python
import pandas as pd

records = [('Author', 'Pham Dinh Khanh', 1993), 
           ('Book', 'ML algorithms to Practice', 2021), 
           ('Target', 'Vi mot cong dong AI vung manh hon', 1), 
           ('No_Donation', 'Community', 2)]

df = pd.DataFrame(data = records, columns = ['contents', 'infos', 'numbers'])
display(df)
```

##  Min, max, mean, meadian, sum

```python
# cut() - phân chia giá trị của một trường liên tục vào những khoảng theo ngưỡng cắt
bins = [-999999, 250, 400, 999999]
labels = ['low', 'normal', 'high']
# low: -999999 <- 250
# normal: 250 <- 400
# high: 400 <- 999999
df['tax_labels'] = pd.cut(df['tax'], bins=bins, labels=labels)
df[df['tax_labels']=='low'].head()
```

```python
# qcut() - phân chia giá trị theo định lượng
import numpy as np

labels = ['low', 'normal', 'high']
tax_labels = pd.qcut(df['tax'], q=3, labels=labels)
np.unique(tax_labels, return_counts = True)
```

```python
# map() - biến 1 giá trị thành 1 giá trị mới dựa trên dictionary mà ta sử dụng 
# giá trị cũ là key , giá trị mới là value
dict_tax =  {
    'low':'thap',
    'normal':'tb',
    'high':' cao'
}
df['tax_labels'].map(dict_tax).head()
```

## Biểu đồ matplotlib in Pandas

```python
#1. biểu đồ line
df1['Age'].plot()
```

```python
#2. Biểu đồ barchart
bins = [-999999, 250, 400, 999999]
labels = ['low', 'normal', 'high']
# low: -999999 <- 250
# normal: 250 <- 400
# high: 400 <- 999999
df['tax_labels'] = pd.cut(df['tax'], bins=bins, labels=labels)
df[df['tax_labels']=='low'].head()
df_sum = df[['tax_labels', 'tax']].groupby('tax_labels').sum().plot.bar()
```

```python
# biểu đồ pie
df_sumary =  df[['tax_labels', 'tax']].groupby('tax_labels').sum()
df_sumary['tax'].plot.pie(autopct = '%1.1f%%')
```

```python
# tự nghịch 1 ví dụ hehe 
bins_age =[25, 26, 27, 28]
label_age = ["younger", "milde", "old"]
df1['labels_age'] = pd.cut(df1['Age'], bins=bins_age, labels=label_age)
df1_sum = df1[['labels_age','Age']].groupby('labels_age').count()
df1_sum.head()
df1_sum['Age'].plot.pie(autopct='%1.2f%%')
```

```python
#3.Biểu đồ boxplot - được sử dụng để quan sát phân phối của biến đối với các giá trị min, max và các ngưỡng phân vị 25%, 50%, 75%. 
# Căn cứ vào boxplot ta có thể biết được khoảng biến thiên của biến rộng hay hẹp, biến phân phối lệch trái hay phải.
df[['tax', 'age', 'b']].boxplot()
```
